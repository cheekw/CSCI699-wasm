```python
chrome = './chrome.json';
safari = './safari.json';
firefox = './firefox.json';
```

```python
# Opening JSON files
data = []
for path in [chrome, safari, firefox]:  
    f = open(path)
    data.append(json.load(f))
```

```python
[chrome_dict, safari_dict, firefox_dict] = data
```

```python
data_by_lib = {}

col_headers = ['Chrome', 'Safari', 'Firefox']  
    
for lib in chrome_dict:
    data_by_lib[lib] = []
    data_by_lib[lib].append(col_headers)
    for func in chrome_dict[lib]:
        row = []
        row.append(func)
        for ms in chrome_dict[lib][func]:
            row.append(chrome_dict[lib][func][ms])
        data_by_lib[lib].append(row)

for lib in safari_dict:
    i = 1;
    for func in safari_dict[lib]:
        time = safari_dict[lib][func]['ms']
        data_by_lib[lib][i].append(time)
        i += 1

for lib in firefox_dict:
    i = 1;
    for func in firefox_dict[lib]:
        time = firefox_dict[lib][func]['ms']
        data_by_lib[lib][i].append(time)
        i += 1

    
data_by_lib
```

```python
title_text = 'Computer Specifications'

for lib in data_by_lib:
    data = data_by_lib[lib]
    column_headers = data.pop(0)
    row_headers = [x.pop(0) for x in data]
    cell_text = []
    for row in data:
        cell_text.append(["{:.2f}".format(x) for x in row])
    plt.figure(linewidth=2,
               tight_layout={'pad':1},
               figsize=(8, 4)
              )
    the_table = plt.table(cellText=cell_text,
                          rowLabels=row_headers,
                          rowLoc='right',
                          colLabels=column_headers,
                          loc='center')
    the_table.scale(1, 1.5)
    cells = the_table.properties()["celld"]

    ax = plt.gca()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.box(on=None)
    # plt.suptitle(title_text)
    # plt.figtext(0.95, 0.05, footer_text, horizontalalignment='right', size=6, weight='light')
    plt.draw()
```
